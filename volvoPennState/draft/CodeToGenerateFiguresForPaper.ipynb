```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

fig, ax = plt.subplots(figsize=(12, 8))
ax.set_xlim(0, 12)
ax.set_ylim(0, 10)
ax.axis('off')

def draw_box(x, y, width, height, label, color, fontsize=10, align='center', weight='bold'):
    box = patches.FancyBboxPatch((x, y), width, height,
                                 boxstyle="round,pad=0.02",
                                 edgecolor="black", facecolor=color, linewidth=1.5)
    ax.add_patch(box)
    
    ha = {'center': 'center', 'left': 'left'}[align]
    x_offset = 0.1 if align == 'left' else width / 2

    ax.text(x + x_offset, y + height / 2, label, fontsize=fontsize,
            ha=ha, va='center', wrap=True, family='monospace', weight=weight if '\n' not in label else 'normal')

# 🔷 Top wide blue box
draw_box(0.8, 7.0, 10.5, 2, "Raw Population Table Row", 'lightblue', fontsize=12)

# 🔴 Left middle box - Dropped
dropped_text = ("Dropped:\n"
                "- VIN\n"
                "- CHASSIS_ID\n"
                "- ENGINE_ASSEMBLY_MONTH\n"
                "- VEH_ASSEMBLY_DATE\n"
                "- VEH_ASSEMBLY_MONTH\n"
                "- INS_DATE\n"
                "- INS_MONTH")
draw_box(0.8, 4.2, 4.5, 2, dropped_text, 'lightcoral', fontsize=9, align='left')

# 🟢 Right middle box - Selected
selected_text = ("Selected Features:\n"
                 "- ENGINE_SIZE, ENGINE_HP, VEH_TYPE\n"
                 "- VEH_ASSEMBLY_DATE (→ vehicle age)\n"
                 "- KOLA_01X ... KOLA_ZMX (588 features)")
draw_box(6.7, 4.2, 4.6, 2, selected_text, 'lightgreen', fontsize=9, align='left')

# 🔽 Arrows from blue box to middle boxes
arrowprops = dict(arrowstyle="->", color='black', linewidth=1.5)
ax.annotate("", xy=(4.0, 6.2), xytext=(4.0, 7.0), arrowprops=arrowprops)  # from Dropped to Top
ax.annotate("", xy=(8.0, 6.2), xytext=(8.0, 7.0), arrowprops=arrowprops)  # from Selected to Top


# 📝 Wrapped explanation text under green box
explanation_text = ("592-Dimensional Build Options Vector\n"
                    "(Used for Clustering Vehicles)")
ax.text(7.0, 3.9, explanation_text, fontsize=9.5, ha='left', va='top', family='monospace', wrap=True)
# Save as PDF
plt.savefig("population_table_flow.pdf", format="pdf", bbox_inches="tight")

# Display the plot
plt.show()


plt.show()
```
